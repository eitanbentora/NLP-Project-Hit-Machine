```python
from scipy import stats
import numpy as np

with open('original_maches.pkl', 'rb') as f:
    original_n_grams_match = pickle.load(f)
    
with open('generated_maches.pkl', 'rb') as f:
    generated_n_grams_match = pickle.load(f)
    
    
alpha = 0.05
print("H0 - generated lyrics are as original as real songs.\n\
H1 - generated lyrics are less original then real songs." )
for ngram in generated_n_grams_match:
    if ngram in original_n_grams_match:
        generated_values = list(generated_n_grams_match[ngram].values())
        original_values = list(original_n_grams_match[ngram].values())
        t_stat, p_value =  stats.ttest_ind(generated_values, original_values)
        if p_value/2 < alpha and np.mean(generated_values) > np.mean(original_values):
            print(f"Null hypothesis rejected for ngram = {ngram}, p-value={p_value:.2f}")
        else:
            print(f"Null hypothesis not rejected for ngram = {ngram}, p-value={p_value:.2f}")
        if t_stat > 0:
            print(f'more shared ngrams of size {ngram} at the generated lyrics')
        else:
            print(f'more shared ngrams of size {ngram} at the original lyrics')
```
